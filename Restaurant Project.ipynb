```python
myOrder = {}

def display_information():
    print("Welcome to PF Chang's!")
    print("Enjoy our selection of appetizers, beverages, entrees, and desserts, crafted with the finest ingredients.\n")
    print("Store Hours and Information:")
    print("Phone Number: (201) 646-1565")
    print("Address: 390 Hackensack Ave, Ste 50, Hackensack, NJ 07601")
    print("Website: https://order.pfchangs.com/menu/9826")
    print("Open Hours:")
    print("  Business: 11 AM - 10 PM")
    print("  Curbside: 11 AM - 10 PM")
    print("  Delivery: 11 AM - 10 PM")
    print("Accepted Cards: Visa, MasterCard, Discover, Amex\n")

def display_appetizer(myOrder):
    appetizers = {
        "Chang's Chicken Lettuce Wraps": 16.29,
        "Edamame": 10.79,
        "Crispy Green Beans": 12.29,
        "Korean Style Cauliflower": 8.59,
        "Dynamite Shrimp": 16.59
    }
    print("Appetizers Menu:")
    for item, price in appetizers.items():
        print(f"{item}: ${price}")
    if input("Would you like to order from this menu? (yes/no): ").strip().lower() == 'yes':
        item = input("What would you like to order? ")
        quantity = int(input("How many would you like to order? "))
        myOrder[item] = {"quantity": quantity, "price": appetizers[item] * quantity}
        print(f"{quantity} of {item} added to your order.")

def display_beverages(myOrder):
    beverages = {
        "Fiji 1L": 7.29,
        "San Pellegrino 1L": 6.59,
        "Pomegranate Lemonade": 6.49,
        "Peach Boba Breeze": 6.49,
        "Strawberry Cucumber Limeade": 6.49
    }
    print("Beverages Menu:")
    for item, price in beverages.items():
        print(f"{item}: ${price}")
    if input("Would you like to order from this menu? (yes/no): ").strip().lower() == 'yes':
        item = input("What would you like to order? ")
        size = input("Select size (small, medium, large): ").strip().lower()
        quantity = int(input("How many would you like to order? "))

        price_adjustment = -1 if size == "small" else 1 if size == "large" else 0
        final_price = (beverages[item] + price_adjustment) * quantity
        myOrder[item] = {"quantity": quantity, "price": final_price}
        print(f"{quantity} of {item} ({size}) added to your order.")

def display_entree(myOrder):
    entrees = {
        "Mongolian Beef": 26.29,
        "Pepper Steak": 23.29,
        "Kung Pao Chicken": 19.59,
        "Orange Chicken": 18.79,
        "Teriyaki Chicken": 18.49
    }
    print("Entrees Menu:")
    for item, price in entrees.items():
        print(f"{item}: ${price}")
    if input("Would you like to order from this menu? (yes/no): ").strip().lower() == 'yes':
        item = input("What would you like to order? ")
        quantity = int(input("How many would you like to order? "))
        myOrder[item] = {"quantity": quantity, "price": entrees[item] * quantity}
        print(f"{quantity} of {item} added to your order.")

def display_desserts(myOrder):
    desserts = {
        "Chocolate Souffle": 10.79,
        "New York-Style Cheesecake": 10.49,
        "The Great Wall of Chocolate": 13.59,
        "Banana Spring Rolls": 11.59,
        "GF Chocolate Souffle": 11.29
    }
    print("Desserts Menu:")
    for item, price in desserts.items():
        print(f"{item}: ${price}")
    if input("Would you like to order from this menu? (yes/no): ").strip().lower() == 'yes':
        item = input("What would you like to order? ")
        quantity = int(input("How many would you like to order? "))
        myOrder[item] = {"quantity": quantity, "price": desserts[item] * quantity}
        print(f"{quantity} of {item} added to your order.")

def modifyOrder(myOrder):
    print("Your current order:")
    for item, details in myOrder.items():
        print(f"{item} (x{details['quantity']}): ${details['price']}")

    if input("Would you like to modify your order? (yes/no): ").strip().lower() == 'yes':
        item_to_modify = input("Enter the item you want to modify: ")
        if item_to_modify in myOrder:
            action = input("Would you like to delete or change quantity? (delete/change): ").strip().lower()
            if action == "delete":
                del myOrder[item_to_modify]
                print(f"{item_to_modify} has been removed.")
            elif action == "change":
                new_quantity = int(input(f"Enter new quantity for {item_to_modify}: "))
                price_per_item = myOrder[item_to_modify]["price"] / myOrder[item_to_modify]["quantity"]
                myOrder[item_to_modify] = {"quantity": new_quantity, "price": price_per_item * new_quantity}
                print(f"Quantity for {item_to_modify} updated to {new_quantity}.")
        else:
            print("Item not found in your order.")

def placeOrder(myOrder):
    subtotal = sum(details["price"] for details in myOrder.values())
    print("\nOrder Summary:")
    for item, details in myOrder.items():
        print(f"{item} (x{details['quantity']}): ${details['price']:.2f}")

    if input("Do you have a coupon? (yes/no): ").strip().lower() == 'yes':
        coupon_type = input("Is it a percentage or a specific amount? (percentage/amount): ").strip().lower()
        if coupon_type == "percentage":
            discount = float(input("Enter discount percentage: ")) / 100
            subtotal -= subtotal * discount
        elif coupon_type == "amount":
            discount = float(input("Enter discount amount: "))
            subtotal -= discount

    tip = float(input("Tip (15%, 18%, or 20%): ").strip('%')) / 100
    total_with_tip = subtotal + (subtotal * tip)

    total_with_tax = total_with_tip * 1.06625

    if input("Would you like delivery? (yes/no): ").strip().lower() == 'yes':
        delivery_charge = 5
        total_with_tax += delivery_charge
        print(f"Delivery charge: ${delivery_charge:.2f}")

    print(f"Total with tax and tip: ${total_with_tax:.2f}")
    print("Thank you for your order!")

def main():
    myOrder = {}
    display_information()
    display_appetizer(myOrder)
    display_beverages(myOrder)
    display_entree(myOrder)
    display_desserts(myOrder)
    modifyOrder(myOrder)
    placeOrder(myOrder)

main()
```
